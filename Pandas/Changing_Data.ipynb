```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

```python
# Creando nuevas columnas, usando sum axis 1 se suma horizontal 0 vertical
# iloc primero la fila, marca todas despues las columnas de la posicion 4 a 10
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df.head()
```

```python
# Mover la columna total al comienzo tiene más sentido
cols = list(df.columns.values)
# Reordenar
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

```python
# Guardando la información nueva
df.to_csv('modified_pkmn_data.csv', index=False)
#df.to_excel()
```
