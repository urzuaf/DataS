```python
import pandas as pd 
df = pd.read_csv('olimpics_dataset/athlete_events.csv')
```

```python
# Contar cantidad de valores por resultado con filtros
df[df['Weight'] < 100]['NOC'].value_counts().head()
```

```python
# Agrupar por pais y calcular el promedio de altura y peso
df.groupby(['NOC'])[['Height', 'Weight']].mean().head()
```

```python
# Agrupando por el NOC, el promedio de la edad y el maximo de peso
df.groupby(['NOC']).agg({'Age': 'mean', 'Weight': 'max'}).head()
```
