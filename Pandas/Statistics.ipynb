```python
import pandas as pd
df = pd.read_csv('modified_pkmn_data.csv')
```

```python
# Agruparemos por tipo y nos dará el total de cada uno
df.groupby(['Type 1']).size()
#print(df.dtypes)

#df.groupby(['Type 2'], dropna=True)[["Total", "Speed"]].mean()

#Filtrar solo las columnas numericas
numeric_columns = df.select_dtypes(include="number")

result = df.groupby("Type 1")[numeric_columns.columns].mean()
result.head()
```

```python
# Que tipo tiene el ataque más alto

df.groupby("Type 1")[['Attack', 'Total']].mean().sort_values('Attack',ascending=False).head()
```

```python
# mostrara el count de cada columna
# No cuenta los datos nulos o vacios
df.groupby("Type 1").count().head()
```
