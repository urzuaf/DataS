```python
import pandas as pd 
athletes = pd.read_csv('olimpics_dataset/athlete_events.csv')
nocs = pd.read_csv('olimpics_dataset/noc_regions.csv')
```

```python
# left outer join, keeping all the athletes data
merged = pd.merge(athletes, nocs, left_on='NOC', right_on='NOC', how='left')

#Rename
merged.rename(columns={'region':'BornCountry'}, inplace=True)
merged.head()
```

```python
usa = merged[merged['BornCountry'] == 'USA'].copy()
gbr = merged[merged['BornCountry'] == 'GBR'].copy()
both = pd.concat([usa,gbr])
both.tail()
```
