```python
import pandas as pd
df = pd.read_csv('coffe.csv')
ath = pd.read_csv('olimpics_dataset/athlete_events.csv')
```

```python
# Añadimos la columna de revenue
df['revenue'] = df['price'] * df['Units Sold']
```

```python
#shift
df['yesteday_revenue'] = df['revenue'].shift(2)
# Porcentaje de cambio en las ventas
df['change'] = df['revenue'] / df['yesteday_revenue'] * 100
df.head(6)
```

```python
#Rankeando a cada atleta por su altura
ath['HeightRank'] = ath['Height'].rank()
ath.sort_values(['HeightRank'],ascending=False).head()
```

```python
# Sumará las columnas, cuidado porque el texto se duplica, entonces solo sumar numeros
#df.select_dtypes('number').cumsum().head()
df['cumulative_revenue'] = df['revenue'].cumsum()
df
```

```python
latte = df[df['Coffee Type'] == 'Latte'].copy()

# Suma acumulada en 3 dias
latte['3day'] = latte['Units Sold'].rolling(3).sum()
latte
```
