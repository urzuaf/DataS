```python
import pandas as pd
df = pd.read_csv('modified_pkmn_data.csv')
```

```python
# Filtering on multiples conditions
# & and | or ~ not
dfg = df.loc[(df["Type 1"] == "Grass") & (df["HP"] >= 70) ]

# Al filtrar los indices se pueden comprometer, para solucionar:
dfg = dfg.reset_index()
dfg.head()
```

```python
# Retorna todo lo NO mega
new_df = df.loc[ ~ df['Name'].str.contains('Mega')]
new_df.head()
```

```python
# REGEX filtering
import re
#new_df = df.loc[df['Type 1'].str.contains('fire|Grass',flags=re.I, regex=True)]
new_df = df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I, regex=True)]
new_df.head()
```

```python
# Cambios condicionales
# Cambia el valor de Fire por Fuego
df.loc[ df["Type 1"] == "Fire" , 'Type 1'] = 'Fuego' 
df
# Cambia el valor de legendario a True en donde el tipo sea 'Fuego'

df.loc[ df["Type 1"] == "Fuego", "Legendary" ] = True
df.head()
```

```python
# Modify multiple columns at the same time

df.loc[df["Total"] > 500, ["Legendary", "Generation"] ] = [True, 99]
df.head(15)
```
